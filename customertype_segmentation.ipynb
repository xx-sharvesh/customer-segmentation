```python
import pandas as pd

# Load the CSV file into a DataFrame
file_path = 'rfm-table.csv'
df = pd.read_csv(file_path, encoding='latin1')

# Define thresholds for categorization
high_frequency_threshold = 7
high_recency_threshold = 100
high_monetary_threshold = 3000

# Function to categorize customers
def categorize_customer(row):
    if row['frequency'] >= high_frequency_threshold and row['recency'] <= high_recency_threshold and row['monetary_value'] >= high_monetary_threshold:
        return 'Active Loyals'
    elif row['monetary_value'] >= high_monetary_threshold and row['frequency'] >= high_frequency_threshold:
        return 'Big Spenders'
    elif row['frequency'] >= high_frequency_threshold and row['recency'] > high_recency_threshold:
        return 'Loyals at Risk'
    elif row['frequency'] < high_frequency_threshold and row['recency'] > high_recency_threshold:
        return 'Churn Candidates'
    elif row['frequency'] == 1:
        return 'One-Time Shoppers'
    elif row['monetary_value'] >= 10000:
        return 'Richie Riches'
    elif row['recency'] > high_recency_threshold and row['frequency'] >= high_frequency_threshold:
        return 'Lost Customers'
    elif row['frequency'] >= high_frequency_threshold:
        return 'Bulk Shoppers'
    else:
        return 'Discount Seekers'

# Apply the categorization function to the DataFrame
df['Customer_Segment'] = df.apply(categorize_customer, axis=1)

# Display the first few rows of the updated DataFrame
df.head(25)
```

```python
# Define email frequency and type based on customer segment
def email_strategy(row):
    if row['Customer_Segment'] == 'Active Loyals':
        return 'Weekly', 'Personalized updates and loyalty rewards'
    elif row['Customer_Segment'] == 'Big Spenders':
        return 'Weekly', 'Premium offers and exclusive deals'
    elif row['Customer_Segment'] == 'Loyals at Risk':
        return 'Bi-weekly', 'Re-engagement campaigns'
    elif row['Customer_Segment'] == 'Churn Candidates':
        return 'Monthly', 'High-impact reactivation emails'
    elif row['Customer_Segment'] == 'One-Time Shoppers':
        return 'Bi-weekly', 'Enticing offers'
    elif row['Customer_Segment'] == 'Richie Riches':
        return 'Monthly', 'Exclusive high-end promotions'
    elif row['Customer_Segment'] == 'Lost Customers':
        return 'Quarterly', 'Strong reactivation offers'
    elif row['Customer_Segment'] == 'Waiting First Order':
        return 'Weekly', 'Reminders and first-order discounts'
    elif row['Customer_Segment'] == 'Discount Seekers':
        return 'Weekly', 'Discount offers'
    elif row['Customer_Segment'] == 'Bulk Shoppers':
        return 'Bi-weekly', 'Bulk purchase deals'
    else:
        return 'Monthly', 'General updates'

# Apply the email strategy function to the DataFrame
df[['Email_Frequency', 'Email_Type']] = df.apply(lambda row: pd.Series(email_strategy(row)), axis=1)

# Display the first few rows of the updated DataFrame
df.head(25)
```

```python
# Save the updated DataFrame to a new CSV file
output_file_path = 'updated_rfm_table.csv'
df.to_csv(output_file_path, index=False)
```
